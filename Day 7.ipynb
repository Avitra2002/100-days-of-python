```python
### Day 7 - Hang Man
import random
word_list= ["ardvark", "baboon", "camel"]
stages=['''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']
ls=[]
mistake=0
switch=False 

## Randomly choose a word
choosen_word= random.choice(word_list)
print(choosen_word)

###making blanks
for i in range(len(choosen_word)):
    ls+= "_"



while mistake<7:
    if "_" in ls:
    
        print(ls)
        user=input("Choose a letter").lower()
        if user in choosen_word:
            ###finding the index of chosen letter
            for idx in range(len(choosen_word)):
                if choosen_word[idx]==user:
                    ls[idx]=user
            print(stages[mistake])
        else:
            mistake+=1
            print("No Match")
            print(stages[mistake])
    else:
        print(ls)
        print("You win")
        break ## exit the loop
print("Game Over")




    


```
